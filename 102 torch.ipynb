```python
import torch

# Create a 2D tensor (2 rows, 3 columns)
x = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])

print("Original shape:", x.shape)   
print("Original tensor:\n", x)

# Use view(-1) to flatten into 1D
y = x.view(-1)

print("\nReshaped shape:", y.shape) 
print("Reshaped tensor:\n", y)
```

```python
x = torch.arange(6)   
y = x.view(2,3) 
z = x.view(3,2)

print(f"x\n {x.shape} \n {x}")
print(f"y\n {y.shape}\n {y}")
print(f"z\n {z.shape}\n {z}")
```

```python
# Original tensor with float values
x = torch.tensor([1.7, 2.9, 3.2], dtype=torch.float32)

print(f"x\n {x.dtype} \n {x}")

# Convert to int32
y = x.to(torch.int32)

print(f"y\n {y.dtype} \n {y}")
```

```python
# x.size() is same as x.shape
x = torch.arange(3,9).reshape(2,3)
print(x)
print(x.size())
print(x.shape)
```

```python
#Squeeze and unsqueeze to remove/add dimension
x = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])
print(f"x\n  {x}")
print(f"Original shape:, {x.shape}")  

# Unsqueeze-- Add a new dimension at position 0
y = x.unsqueeze(0)
print(f"y\n  {y}")
print(f"After unsqueeze(0):, {y.shape}")  

# Squeeze-- Remove that dimension again
z = y.squeeze(0)
print(f"z\n  {z}")
print(f"After squeeze(0):, {z.shape}")    # torch.Size([2, 3])
```

```python
#Transpose
x = torch.tensor([[1, 2, 3],
                  [4, 5, 6]])
print(f"Original tensor x:\n{x}")
print(f"Original shape : {x.shape}\n") 

y = x.transpose(0,1)

print(f"Transposed tensor y:\n{y}")
print(f"\nTransposed shape: {y.shape}")  
```

```python
#Higher Dimensions -Transpose

x = torch.randn(2, 3, 4)   


print(f"Original tensor x:\n{x}")
print(f"\nOriginal shape : {x.shape}\n")

y = x.transpose(0,1)       # swap first two dimensions

print(f"\nTransposed shape: {y.shape}") 
```

```python
#permute
x = torch.randn(2, 3, 4)   
y = x.permute(2, 0, 1)     # reorder dimensions

print(x.shape)  
print(y.shape)  
```

## transpose vs permute

| Function        | Syntax                        | What it does                  | Example result          |
|-----------------|-------------------------------|-------------------------------|-------------------------|
| transpose(a,b)  | `x.transpose(dim0, dim1)`     | Swaps two dimensions only     | (2,3,4) → (3,2,4)       |
| permute(...)    | `x.permute(dims...)`          | Reorders all dimensions at once | (1,2,3,4) → (1,4,3,2) |


```python
### Math & Statistics

x = torch.tensor([1.0, 2.0, 3.0, 4.0])

print(f"tensor x: {x}")
# 1. Max / Min values
print(f"\ntorch.max(x):{torch.max(x)}")   
print(f"\ntorch.min(x):{torch.min(x)}")   

# 2. Index of Max / Min
print(f"\ntorch.argmax(x):{torch.argmax(x)}")  
print(f"\ntorch.argmin(x):{torch.argmin(x)}")  

# 3. Standard Deviation
print(f"\ntorch.std(x):{torch.std(x)}")  

# 4. Variance
print(f"\ntorch.var(x):{torch.var(x)}")  

# 5. Vector Norm (magnitude)
# sqrt(1^2 + 2^2 + 3^2 + 4^2) = sqrt(30) ≈ 5.477
print(f"\ntorch.norm(x):{torch.norm(x)}")  

# 6. Clamp values to a range
y = torch.tensor([-2.0, 0.5, 1.5, 3.0])
print(f"\ntensor y: {y}")
print(f"\ntorch.clamp(y, min=0, max=1): {torch.clamp(y, min=0, max=1)}")
```

```python
x = torch.tensor([1.0, 2.0, 3.0])
print(f"tensor x: {x}\n")
print("Power:", torch.pow(x, 2))     
print("Square root:", torch.sqrt(x)) # tensor([1.0000, 1.4142, 1.7321])
print("Exponential:", torch.exp(x))  # e^x
print("Logarithm:", torch.log(x))    # natural log
```

```python
angles = torch.tensor([0.0, 3.1416/2, 3.1416])  # 0, π/2, π
print(f"tensor angles: {angles}\n")

print("sin:", torch.sin(angles))   # tensor([0., 1., 0.])
print("cos:", torch.cos(angles))   # tensor([1., 0., -1.])
print("tan:", torch.tan(angles))   # tensor([0., huge, 0.])
```

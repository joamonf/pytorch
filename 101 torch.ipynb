## Dot Product

    The dot product is defined for two vectors (1D tensors).

    It multiplies corresponding elements and sums them up.

```python
u = torch.tensor([1, 2, 3])
v = torch.tensor([4, 5, 6])

dot = torch.dot(u, v)
print(dot)   # 1*4 + 2*5 + 3*6 = 32
```

## Matrix Multiplication (torch.matmul)

    Works for 2D tensors (matrices) or higher dimensions.

    It’s a generalization of the dot product: each element in the result is a dot product between a row of the first matrix and a column of the second.
    
    If you have two tensors with shapes,(A:m×n)and(B:n×p)then: torch.matmul(A,B)⇒(m×p) and, inner dimensions (n) must match.

```python
a = torch.tensor([[1, 2], [3, 4]])
b = torch.tensor([[2, 0], [1, 2]])

result = torch.matmul(a, b)
print(f"a \n {a}")
print(f"b \n {b}")
print(f"matmul \n {result}")
```
